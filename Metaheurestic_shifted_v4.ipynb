```python
class shifted_rosenbrock_problem:
    def __init__(self, dim):
        self.dim = dim
    def fitness(self, x):
        sum_val = np.zeros((1,))
        fitness_val = np.zeros((1,))
        for i in range(len(x)-1):
            sum_val[0] = sum_val[0] + 100*(((x[i]-rosenbrock[i]+1)**2-(x[i+1]-rosenbrock[i+1]+1))**2)+ (x[i]-rosenbrock[i])**2
        fitness_val[0] = sum_val[0] + f_bias[2]
        return fitness_val
    def get_bounds(self):
        return (np.full((self.dim,),-100),np.full((self.dim,),100))
    def get_name(self):
        return "Shifted Schwefel Function"
    def get_extra_info(self):
        return "\tDimensions: " + str(self.dim)
```

```python
class shifted_rastrigin_problem:
    def __init__(self, dim):
        self.dim = dim
    def fitness(self, x):
        sum_val = np.zeros((1,))
        fitness_val = np.zeros((1,))
        for i in range(len(x)):
            sum_val[0] = sum_val[0] + (x[i]-rastrigin[i])**2 + 10 * (np.cos(2*np.pi*(x[i]-rastrigin[i])))+10
        fitness_val[0] = sum_val[0] + f_bias[3]
        return fitness_val
    def get_bounds(self):
        return (np.full((self.dim,),-5),np.full((self.dim,),5))
    def get_name(self):
        return "Shifted Rastrigin Function"
    def get_extra_info(self):
        return "\tDimensions: " + str(self.dim)
```

```python
class shifted_griewank_problem:
    def __init__(self, dim):
        self.dim = dim
    def fitness(self, x):
        sum_val = np.zeros((1,))
        prod_val = np.ones((1,))
        fitness_val = np.zeros((1,))
        for i in range(len(x)):
            prod_val[0] = prod_val[0] * np.cos((x[i]- griewank[i])/np.sqrt(i+1))
            sum_val[0] = sum_val[0] + np.power((x[i]-griewank[i]),2)/4000
        fitness_val[0] = sum_val[0] - prod_val[0] + f_bias[4] +1
        return fitness_val
    def get_bounds(self):
        return (np.full((self.dim,),-600),np.full((self.dim,),600))
    def get_name(self):
        return "Shifted Griewank Function"
    def get_extra_info(self):
        return "\tDimensions: " + str(self.dim)
```

```python
class shifted_ackley_problem:
    def __init__(self, dim):
        self.dim = dim
    def fitness(self, x):
        sum_val_1 = np.zeros((1,))
        sum_val_2 = np.zeros((1,))
        fitness_val = np.zeros((1,))
        D = len(x)
        for i in range(len(x)):
            z = x[i]-ackley[i]
            sum_val_1[0] = sum_val_1[0] + z**2
            sum_val_2[0] = sum_val_2[0] + np.cos(2*np.pi*z)
        fitness_val[0] = (-20)*np.exp(-0.2 * np.sqrt(sum_val_1[0])/D) - np.exp(sum_val_2[0]/D) + 20 + 150 + f_bias[5]
        return fitness_val
    def get_bounds(self):
        return (np.full((self.dim,),-32),np.full((self.dim,),32))
    def get_name(self):
        return "Shifted Ackley Function"
    def get_extra_info(self):
        return "\tDimensions: " + str(self.dim)
```

```python
def create_algo_obj(algo_type,algo_gen):
    if algo_type=="de":
        algo_obj = pg.algorithm(pg.de(gen = algo_gen))
    elif algo_type=="pso":
        algo_obj = pg.algorithm(pg.pso(gen = algo_gen))
    elif algo_type=="bee_colony":
        algo_obj = pg.algorithm(pg.bee_colony(gen = algo_gen))
    elif algo_type=="sga":
        algo_obj = pg.algorithm(pg.sga(gen = algo_gen))
    elif algo_type=="cmaes":
        algo_obj = pg.algorithm(pg.cmaes(gen = algo_gen))
    elif algo_type=="sa":
        algo_obj = pg.algorithm(pg.simulated_annealing())
    return algo_obj
```

```python
def create_prob(prob_variant,prob_dim):
    if prob_variant == 'shifted_sphere':
        prob = pg.problem(shifted_sphere_problem(dim = prob_dim))
    elif prob_variant == 'shifted_schwefel':
        prob = pg.problem(shifted_schwefel_problem(dim = prob_dim))
    elif prob_variant == 'shifted_rosenbrock':
        prob = pg.problem(shifted_rosenbrock_problem(dim = prob_dim))
    elif prob_variant == 'shifted_rastrigin':
        prob = pg.problem(shifted_rastrigin_problem(dim = prob_dim))
    elif prob_variant == 'shifted_griewank':
        prob = pg.problem(shifted_griewank_problem(dim = prob_dim))
    elif prob_variant == 'shifted_ackley':
        prob = pg.problem(shifted_ackley_problem(dim = prob_dim))
    return prob
```

```python
def calc_fitness(iteration,algo_type,algo_gen,problem_type,pg_population_size,problem_dimension=50):
    fitness_val = []
    pop_obj = []
    prob_obj = create_prob(problem_type,problem_dimension)
    pop = pg.population(prob_obj,size = pg_population_size,seed=123)
    algo_obj = create_algo_obj(algo_type,algo_gen)
    flag = 0
    for i in range(iteration):
        fitness_val.append(pop.champion_f)
        pop_obj.append(pop)
        pop = algo_obj.evolve(pop)

    return (fitness_val,pop_obj)
```
